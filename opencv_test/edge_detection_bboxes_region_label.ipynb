```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def showImg(img):
    plt.xticks([]), plt.yticks([])
    plt.imshow(img, 'gray')
#     cv2.imshow('image', img)
#     cv2.waitKey(0)
#     cv2.destroyAllWindows()
```

```python
img = cv2.imread('mainlogo.png', 0)
showImg(img)
```

```python
ret, mask = cv2.threshold(img, 220, 255, cv2.THRESH_BINARY_INV)
showImg(mask)
```

```python
kernel = np.ones((5,5), np.uint8)
erosion = cv2.erode(mask, kernel, iterations=1)
dilation = cv2.dilate(mask, kernel, iterations=1)
# plt.figure(figsize=(15,9))
plt.subplot(131)
showImg(mask)
plt.subplot(132)
showImg(erosion)
plt.subplot(133)
showImg(dilation)
```

```python
opening = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)
closing = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)
plt.subplot(121)
showImg(opening)
plt.subplot(122)
showImg(closing)
```

```python
plt.subplot(131)
showImg(cv2.subtract(dilation, mask))
plt.subplot(132)
showImg(cv2.subtract(mask, erosion))
plt.subplot(133)
showImg(cv2.subtract(dilation, erosion))
```

```python
img = cv2.imread('sudoku.jpg',0)

laplacian = cv2.Laplacian(img,cv2.CV_64F)
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)

plt.subplots(figsize=(20,10))
plt.subplot(2,2,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2),plt.imshow(laplacian,cmap = 'gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,3),plt.imshow(sobelx,cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,4),plt.imshow(sobely,cmap = 'gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])
```

```python
img = cv2.imread('kaptcha.jpg', 1)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#detect regions in gray scale image
mser = cv2.MSER_create()
regions, bboxes = mser.detectRegions(gray)
digits = []
for b in bboxes:
    x,y,w,h = b
    if w>=13 and w<=16 and h==22:
        digits += [(x,y,w,h)]
#         cv2.rectangle(img, (x, y), (x + w, y + h), (255, 255, 0), 1)

# plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB));
```

```python
img = cv2.imread('coins.jpg', 0)
plt.imshow(img, 'gray');
```

```python
ret, thresh = cv2.threshold(img,0,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
plt.imshow(thresh, 'gray');
```

```python
kernel = np.ones((3,3), np.uint8)
closing = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel, iterations=1)
plt.imshow(closing, 'gray');
```

```python
opening = cv2.morphologyEx(thresh,cv2.MORPH_OPEN,kernel, iterations = 2)
# sure background area
sure_bg = cv2.dilate(opening,kernel,iterations=3)

# # Finding sure foreground area
dist_transform = cv2.distanceTransform(opening,cv2.DIST_L2,5)
ret, sure_fg = cv2.threshold(dist_transform,0.7*dist_transform.max(),255,0)
sure_fg = np.uint8(sure_fg)
unknown = cv2.subtract(sure_bg,sure_fg)
plt.imshow(sure_fg, 'gray');
```

```python
a = cv2.erode(thresh, kernel, iterations=11)
unknown = cv2.subtract(thresh, a)
plt.imshow(a, 'gray');
```

```python
ret, markers = cv2.connectedComponents(a)
markers = markers+1
markers[unknown==255] = 0
```

```python
imgcolor = cv2.imread('coins.jpg', 1)
markers = cv2.watershed(imgcolor,markers)
imgcolor[markers == -1] = [255,0,0]
plt.imshow(markers);
```
