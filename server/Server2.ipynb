```python
import socket
from PIL import Image
import io
import sys
```

```python
sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
host = '0.0.0.0'
port = 5555
sock.bind((host, port))
sock.listen()
```

```python
(client_sock, client_addr) = sock.accept()
print("Client Info: ", client_sock, client_addr)
```

```python
msg = client_sock.recv(1024000)
```

```python
image = Image.open(io.BytesIO(msg))
image
```

```python
sys.getsizeof(msg)
```

```python
sock.close()
```
